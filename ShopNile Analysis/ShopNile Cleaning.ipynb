# **Data Cleaning**

## Customers DF Cleaning

### Detect & Remove Duplicates 

```python
Customers_Duplicates_ratio = Dim_Customers_df.duplicated().mean() * 100
print(f"Duplicates Percentage: {Customers_Duplicates_ratio:.2f}%")
Dim_Customers_df = Dim_Customers_df.drop_duplicates()
```

### Detecte & Repalce Outliers on Age

```python
Q1 = Dim_Customers_df["age"].quantile(0.25)
Q3 = Dim_Customers_df["age"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
Outliers = (Dim_Customers_df["age"] < lower_bound) | (Dim_Customers_df["age"] > upper_bound)
outlier_ratio = Outliers.mean() * 100
print(f"Age Outliers Percentage: {outlier_ratio:.2f}%")
median_age = Dim_Customers_df["age"].median()
Dim_Customers_df.loc[Outliers, "age"] = int(median_age)
```

### Exploring Age Distribution

```python
Dim_Customers_df["age"].plot(kind="hist",title="Age Distribution")
plt.xlabel("Age")
plt.show()
```

### Detecte & Repalce Outliers & Missing Values on Income

```python
Q1 = Dim_Customers_df["income"].quantile(0.25)
Q3 = Dim_Customers_df["income"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
Outliers = (Dim_Customers_df["income"] < lower_bound) | (Dim_Customers_df["income"] > upper_bound)
outlier_ratio = Outliers.mean() * 100
print(f"Income Outliers Percentage: {outlier_ratio:.2f}%")
median_income = Dim_Customers_df["income"].median()
Dim_Customers_df.loc[Outliers, "income"] = median_income
Income_Nulls = Dim_Customers_df["income"].isna().mean() * 100
print(f"Income Nulls Percentage: {Income_Nulls:.2f}%")
Dim_Customers_df["income"] = Dim_Customers_df["income"].fillna(median_income)
```

### Exploring Income Distibution

```python
Dim_Customers_df["income"].plot(kind="hist",title="Income Distribution")
plt.xlabel("Income")
plt.show()
```

### Fix City Errors & Replace Missing Values

```python
Dim_Customers_df["city"] = Dim_Customers_df["city"].str.capitalize()
Dim_Customers_df["city"].replace("Alex","Alexandria",inplace=True)
City_Nulls = Dim_Customers_df["city"].isna().mean() * 100
print(f"City Nulls Percentage: {City_Nulls:.2f}%")
Dim_Customers_df["city"] = Dim_Customers_df["city"].fillna("Unknown")
```

### Exploring Customers_City

```python
Dim_Customers_df["city"].value_counts().plot(kind="bar",title="Customers by City")
plt.xlabel("City")
plt.show()
```

### Fix Signup_Date Data Type

```python
Dim_Customers_df["signup_date"] = pd.to_datetime(Dim_Customers_df["signup_date"])
```

## Products DF Cleaning

### Detect & Replace Outliers & Missung Values on Price

```python
Q1 = Dim_Products_df["price"].quantile(0.25)
Q3 = Dim_Products_df["price"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
Outliers = (Dim_Products_df["price"] < lower_bound) | (Dim_Products_df["price"] > upper_bound)
outlier_ratio = Outliers.mean() * 100
print(f"Price Outliers Percentage: {outlier_ratio:.2f}%")
median_price = Dim_Products_df["price"].median()
Dim_Products_df.loc[Outliers, "price"] = median_price
Price_Nulls = Dim_Products_df["price"].isna().mean() * 100
print(f"Price Nulls Percentage: {Price_Nulls:.2f}%")
Dim_Products_df["price"] = Dim_Products_df["price"].fillna(median_price)
```

## Orders DF Cleaning

### Detect & Remove Duplicates

```python
Orders_Duplicates_ratio = Fact_Orders_df.duplicated().mean() * 100
print(f"Duplicates Percentage: {Orders_Duplicates_ratio:.2f}%")
Fact_Orders_df = Fact_Orders_df.drop_duplicates()
```

### Checking Invalid IDs

```python
invalid_customer_ids = ~Fact_Orders_df["customer_id"].isin(Dim_Customers_df["customer_id"])
invalid_product_ids = ~Fact_Orders_df["product_id"].isin(Dim_Products_df["product_id"])
print(f"Invalid Customer IDs Percentage: {invalid_customer_ids.mean() * 100:.2f}%")
print(f"Invalid Product IDs Percentage: {invalid_product_ids.mean() * 100:.2f}%")
```

### Replaace Missing Values on Discount

```python
Discount_Nulls = Fact_Orders_df["discount"].isna().mean() * 100
print(f"Discount Nulls Percentage: {Discount_Nulls:.2f}%")
median_discount = Fact_Orders_df["discount"].median()
Fact_Orders_df["discount"] = Fact_Orders_df["discount"].fillna(median_discount)
```

### Exploring Discount Distribution

```python
Fact_Orders_df["discount"].plot(kind="hist",title="Discount Distribution")
plt.xlabel("Discount")  
plt.show()
```

### Fix Payment_Method Errors and Replace Missing Values

```python
Fact_Orders_df["payment_method"] = Fact_Orders_df["payment_method"].str.title()
Fact_Orders_df["payment_method"].replace("Card","Credit Card",inplace=True)
Payment_Method_Nulls = Fact_Orders_df["payment_method"].isna().mean() * 100
print(f"Payment Method Nulls Percentage: {Payment_Method_Nulls:.2f}%")
Fact_Orders_df["payment_method"] = Fact_Orders_df["payment_method"].fillna("Unknown")
```

### Exploring Payment_Method

```python
Fact_Orders_df["payment_method"].value_counts().plot(kind="bar",title="Orders by Payment Method")
plt.xlabel("Payment Method")
plt.show()
```

### Fix Order_Date Data Type

```python
Fact_Orders_df["order_date"] = pd.to_datetime(Fact_Orders_df["order_date"])
```

## Payments DF Cleaning

### Checking Invalid ID

```python
invalid_order_ids = ~Fact_Payments_df["order_id"].isin(Fact_Orders_df["order_id"])
print(f"Invalid Order IDs Percentage: {invalid_order_ids.mean() * 100:.2f}%")
```

### Detect & Replace Outliers & Nulls & Negative on Amount_paid

```python
Q1 = Fact_Payments_df["amount_paid"].quantile(0.25)
Q3 = Fact_Payments_df["amount_paid"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
Outliers = (Fact_Payments_df["amount_paid"] < lower_bound) | (Fact_Payments_df["amount_paid"] > upper_bound)
outlier_ratio = Outliers.mean() * 100
print(f"Amount paid Outliers Percentage: {outlier_ratio:.2f}%")
median_amount_paid = Fact_Payments_df["amount_paid"].median()
Fact_Payments_df.loc[Outliers, "amount_paid"] = median_amount_paid
Payment_Amont_Nulls = Fact_Payments_df["amount_paid"].isna().mean() * 100
print(f"Amount Paid Nulls Percentage: {Payment_Amont_Nulls:.2f}%")
Fact_Payments_df["amount_paid"] = Fact_Payments_df["amount_paid"].fillna(median_amount_paid)
invalid_amount = Fact_Payments_df["amount_paid"] < 0
invalid_amount_ratio = invalid_amount.mean() * 100
print(f"Invalid Amount Paid Percentage: {invalid_amount_ratio:.2f}%")
Fact_Payments_df["amount_paid"] = Fact_Payments_df["amount_paid"].abs()
```

### Exploring Amount_Paid Distribution

```python
Fact_Payments_df["amount_paid"].plot(kind="hist",title="Amount_Paid Distribution")
plt.xlabel("Amount_Paid")
plt.show()
```

### Fix Payment_Status Errors

```python
Fact_Payments_df["payment_status"] = Fact_Payments_df["payment_status"].str.capitalize()
```

### Exploring Payment_status

```python
Fact_Payments_df["payment_status"].value_counts().plot(kind="bar",title="Orders by Payment Status")
plt.xlabel("Payment Status")
plt.show()
```

### Fix Payment_Date Data Type

```python
Fact_Payments_df["payment_date"] = pd.to_datetime(Fact_Payments_df["payment_date"])
```
